
Data has no null values.

```python
#check for duplicates

duplicates = data[data['customerID'].duplicated(keep=False)]

if not duplicates.empty:
    print("Duplicated customer IDs:")
    print(duplicates['customerID'].unique())
else:
    print("There are no duplicates in the 'customerID' column.")
```

The data does not have duplicated clients.
We proceed check the datatypes accuracy for each column.

```python
#change Totalcharges datatype to float

data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
```

For columns Partner, Dependents, PhoneService, MultipleLines, InternetService, OnlineSecurity, OnlineBackup, DeviceProtection, TechSupport, StreamingTV, StreamingMovies I will convert the datatypes to boolen. The first step is to convert columns values to Yes & No, then convert the data to 1 & 0 for better manipulation using powerBI.

```python
#unique values in columns

columns_to_check = ["MultipleLines", "OnlineSecurity", "OnlineBackup", "DeviceProtection", "TechSupport", "StreamingTV", "StreamingMovies"]

for column in columns_to_check:
    unique_values = data[column].unique()
    print(f"Unique values in {column}:")
    print(unique_values)
    print()
```

```python
#Replacing the values with "NO"

data['MultipleLines'] = data['MultipleLines'].replace('No phone service', 'No')

columns_to_replace = ["OnlineSecurity", "OnlineBackup", "DeviceProtection", "TechSupport", "StreamingTV", "StreamingMovies"]

for column in columns_to_replace:
    data[column] = data[column].replace('No internet service', 'No')
```

```python
#replace column values 'Ýes' & 'No' with 1 & 0

columns_to_replace = ["Partner", "Dependents", "PhoneService", "MultipleLines", "InternetService",
                      "OnlineSecurity", "OnlineBackup", "DeviceProtection", "TechSupport",
                      "StreamingTV", "StreamingMovies", "PaperlessBilling", "Churn"]

replacement_mapping = {'Yes': 1, 'No': 0}

data[columns_to_replace] = data[columns_to_replace].replace(replacement_mapping)
```

```python
# Convert the columns to boolean

data[columns_to_replace] = data[columns_to_replace].astype(bool)
```

```python
#export data to excel for visualization in PowerBI

data.to_excel('telecom.xlsx', index=False)
```
